```python
from dataset_management import get_dataset
import pandas as pd
```

```python
ds=get_dataset(dataset_name="fig5_full")
anonymizer={}
for b_n,bench in enumerate(ds.benchmark.unique()):
    anonymizer[bench]=f"B-{b_n}"
for a_n,alg in enumerate(ds.algorithm.unique()):
    anonymizer[alg]=f"A-{a_n}"
for b_n,bench in enumerate(ds.bench_prior.unique()):
    anonymizer[bench]=f"{anonymizer[bench.rsplit('_',1)[0]]}_{bench.rsplit('_',1)[1]}"
print(anonymizer)

def get_anonymized_name(row):
    return pd.Series([anonymizer[row["benchmark"]],anonymizer[row['algorithm']],anonymizer[row["bench_prior"]]])


ds[['benchmark','algorithm','bench_prior']]=ds.apply(get_anonymized_name,axis=1)
print(ds)
```
